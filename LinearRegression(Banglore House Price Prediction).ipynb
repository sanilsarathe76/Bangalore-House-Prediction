```python
# Load Libraries
import pandas as pd
import numpy as np
```

# Apply One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse = False)
```

```python
oh_enc_arr = oh_enc.fit_transform(df3[['location','size','total_sqft']])
oh_enc_arr
```

```python
dummies.keys()
```

```python
# convert into dataframe
oh_enc_df = pd.DataFrame(dummies)
oh_enc_df
```

```python
X = oh_enc_df.drop('price', axis=1)
y = oh_enc_df['price']
 
print('Shape of X = ', X.shape)
print('Shape of y = ', y.shape)
```

## Splitting the DataSet Into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 51)
print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train) 
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Linear Regression - ML Model Training
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

## Predict the value of Home and Test

```python
X_test[0, :]
```

```python
lr.predict([X_test[0, :]])
```

```python
lr.predict(X_test)
```

```python
y_test
```

```python
# compare Accuracy
lr.score(X_test, y_test)
```

# Implementing Ridge and Lasso Regression
## Ridge and Lasso is used for Improve accuracy of our model.

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
ls = Lasso()

ls.fit(X_train,y_train)

ls.score(X_test, y_test)
```

```python
# Increase the value of alpha for improving accuracy

rd2 = Ridge(alpha = 2)

rd2.fit(X_train,y_train)

rd2.score(X_test, y_test)
```

```python
# Increase the value of alpha for improving accuracy

ls2 = Lasso(alpha = 2)

ls2.fit(X_train,y_train)

ls2.score(X_test, y_test)
```

# Model Evaluation
# Root Mean Square Error( Is used for calculating error in our model.)

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
mse =  mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)

print('MSE = ', mse)
print('RMSE = ', rmse)
```
